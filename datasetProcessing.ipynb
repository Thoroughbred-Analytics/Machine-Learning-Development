```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from dataHandler import * 
```

```python
df = pd.read_csv('data/horseData.csv')
df.head()
```

```python
df = clean_data(df)
df.head()
```

Getting the form, rating, rawERg from the father and mother horse
- Can also add the parent's erg as well

```python
df['avgSireRating'] = (df.groupby('sire')['rating'].transform('mean'))
df['avgSireErg'] = (df.groupby('sire')['erg'].transform('mean'))
df['avgSireForm'] = (df.groupby('sire')['form'].transform('mean'))

df['avgDamRating'] = (df.groupby('dam')['rating'].transform('mean'))
df['avgDamErg'] = (df.groupby('dam')['erg'].transform('mean'))

df['avgBmSireRating'] = (df.groupby('bmSire')['rating'].transform('mean'))
df['avgBmSireErg'] = (df.groupby('bmSire')['erg'].transform('mean'))
df['avgBmSireForm'] = (df.groupby('bmSire')['form'].transform('mean'))

df.to_csv('data/encodedHorseData.csv')
```
